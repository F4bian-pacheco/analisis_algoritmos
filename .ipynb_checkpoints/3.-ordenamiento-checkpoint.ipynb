# Ordenamiento

El ordenamiento de arreglos corresponde a encontrar una permutaci√≥n tal que cada elemento sea menor a su sucesor. 

```python
def selection_sort(collection):
    length = len(collection)
    for i in range(length):
        least = i
        for k in range(i + 1, length):
            if collection[k] < collection[least]:
                least = k
        collection[least], collection[i] = (collection[i], collection[least])
    return collection

def insertion_sort(collection):
    for index in range(1, len(collection)):
        while 0 < index and collection[index] < collection[index - 1]:
            collection[index], collection[index - 1] = collection[index - 1], collection[index]
            index -= 1
    return collection
```

```python
import numpy as np
import time

def gen_random(size):
    return np.random.permutation(np.arange(size,dtype=np.int8))

def gen_worst_case(size):
    return np.arange(size,-1,-1,dtype=np.int8)

def gen_best_case(size):
    return np.arange(size,dtype=np.int8)

arr=gen_best_case(10000)

start=time.clock()
insertion_sort(arr.copy())
end=time.clock()
t1=end-start

start=time.clock()
selection_sort(arr.copy())
end=time.clock()
t2=end-start



print('Insertion Sort: %2f'%t1)
print('Selection Sort : %2f'%t2)
```

```python
def merge_sort(A):
    n=len(A)
    if len(A) <= 1:
      return A
    L = mergesort(A[0:n/2])
    R = mergesort(A[n/2:n])
    return merge(L, R)

def merge(L, R):
    result = []
    l_idx, r_idx = (0, 0)
    while l_idx < len(L) and r_idx < len(R):
      if L[l_idx] < R[r_idx]:
        result.append(L[l_idx])
        l_idx += 1
      else:
        result.append(R[r_idx])
        r_idx += 1
    result.extend(L[l_idx:len(L)])
    result.extend(R[r_idx:len(R)])
    return result
```

```python
arr=gen_random(10)

sorted_arr=mergesort(arr)

print sorted_arr
```

```python
import pandas as pd
import sys

method=[mergesort,selection_sort,insertion_sort]
fun=[gen_random,gen_worst_case,gen_best_case]
n_data=[1000,2000,3000,4000,5000]
n_samp=20

s_n=[]
t_i=[]
t_s=[]
t_m=[]

for i in n_data:
    for j in range(n_samp):
        s_n.append(i)
        arr=gen_random(i)
        # insertion sort
        start=time.clock()
        insertion_sort(arr.copy())
        end=time.clock()
        t_i.append(end-start)
        # selection sort
        start=time.clock()
        selection_sort(arr.copy())
        end=time.clock()
        t_s.append(end-start)
        # merge sort
        start=time.clock()
        merge_sort(arr.copy())
        end=time.clock()
        t_m.append(end-start) 
        
```

```python
data_insertion={'n':s_n,'time':t_i,'method':['insertion_sort']*len(t_i)}
df_insertion=pd.DataFrame(data_insertion)
data_selection={'n':s_n,'time':t_s,'method':['selection_sort']*len(t_s)}
df_selection=pd.DataFrame(data_selection)
data_merge={'n':s_n,'time':t_m,'method':['merge_sort']*len(t_m)}
df_merge=pd.DataFrame(data_merge)
df=pd.concat([df_selection,df_insertion,df_merge],axis=0)
df.head()
```

```python
import numpy as np
import statsmodels.formula.api as sm

def get_timing_series(data, quadratic=True):
    if quadratic:
        data['n_squared'] = data.n**2
        model = sm.ols('time ~ n_squared', data=data).fit()
        predictions = [model.params.dot([1.0, i, i**2]) for i in s_n]
        return pd.Series(predictions, index=pd.Index(s_n))
    else: # assume n log(n)
        data['nlogn'] = data.n * np.log(data.n)
        model = sm.ols('n ~ nlogn', data=data).fit()
        predictions = [model.params.dot([1.0, i, i*np.log(i)]) for i in s_n]
        return pd.Series(predictions, index=pd.Index(s_n))

```
